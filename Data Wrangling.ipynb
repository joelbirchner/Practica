### Si quisiera que el filtro cumpla dos condiciones, podria usar los operadores & (and) y | (or)

```python
# Ejemplo en donde deben cumplirse ambas condiciones
data3 = data[(data['Day Mins']>300) & (data['State']=='NY')] # Cada condicion va encerrada en parentesis
data3.shape # Me dice cuales son las filas que cumplen ambas condiciones. El seg numero indica las columnas. 
# Puedo usar varios &
```

```python
# Ejemplo en donde se cumple una condicion o la otra
data4 = data[(data['Day Mins']>300) | (data['State']=='NY')]
data4.shape
```

### Puedo filtrar comparando los valores de dos columnas

```python
data5 = data[data['Day Calls'] < data['Night Calls']] #Se usan corchetes igual que cuando trabajo con ==
data5.shape
```

```python
import numpy as np
```

```python
# Numeros enteros aleatorios en un intervalo
np.random.randint(1,100)
```

```python
# Numeros decimales aleatorios entre 0 y 1, es el que mas se usa
np.random.random()
```

```python
# Funcion que genera una lista de n numeros enteros aleatorios dentro un intervalo (a,b)
def randint_list(n, a, b):
    x = []
    for i in range (n):
        x.append(np.random.randint(a,b))
    return x
```

```python
randint_list(25,0,50)
```

## Seed
Lo que hace esta semilla es permitirme obtener los mismos numeros aleatorios en el futuro para reproducir los resultados obtenidos durante la fase de analisis, al momento de mostrar la informacion obtenida.
Me permite reproducir el experimento aleatorio.

```python
np.random.seed(2018) #Este numero lo elijo yo. Y lo mantengo.
for i in range(5):
    print(np.random.random())
```
