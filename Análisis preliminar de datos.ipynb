## Faltante de datos
Eso muestra un listado de los valores faltantes representados con True.
Ahora tengo que sumarlos:

```python
pd.isnull(data["body"]).ravel().sum()
```

### Para borrar filas que contengan faltantes de datos
Se usa la funcion dropna(axis=0 #es para trabajar borrando filas, el 1 es para borrar columnas)
El segundo argumento es how="all" #solo borra si en todas las columnas hay faltantes;
how="any" borra la fila si hay algun faltante en la misma.

```python
data2=data
```

```python
data2.dropna(axis = 0, how = 'any')
```

### Rellenar faltantes con algun dato
Fillna no sobrescribe la tabla, solo reemplaza y lo muestra.
Para que si pueda hacer esto necesito poner data2['xx'] = data2['xx'].fillna()

```python
data3 = data
```

```python
data3['body'] = data3['body'].fillna(0)
data3['home.dest'] = data3['home.dest'].fillna('Desconocido')
data3.tail()
```

```python
pd.isnull(data3['age']).ravel().sum()
```

```python
data3['age'].fillna(data3['age'].mean())
#Completa los faltantes con el promedio de la columna edad
```

```python
# Otra forma de completar es colocando el metodo forward fill al fillna asi:
data3['age'].fillna(method='ffill')
```

```python
# O bien con el metodo backward fill que toma un valor siguiente al nan y lo coloca en este
data3['age'].fillna(method = 'bfill')
```

```python
data = data.drop(['sex'], axis = 1) #se coloca el nombre de la columna entre corchetes para eliminar el array que contiene a esa 
#columna del dataset. Debe especificarse axis = 1 para que borre la columna y no una fila.
#Hay que colocar data = antes de usar este metodo para que se modifique el dataset. Sino esto queda aqui colgado (parece).
```

```python
# Ahora hay que concatenar el dataframe de data con la variable dummy
data = pd.concat([data, dummy_sex], axis = 1)
data
```

```python
def createDummies(df, var_name):
    dummy = pd.get_dummies(df[var_name], prefix = var_name)
    df = df.drop(var_name, axis = 1)
    df = pd.concat([df,dummy], axis = 1)
    return df
    
```
