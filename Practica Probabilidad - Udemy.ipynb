```python
import matplotlib.pyplot as plt
n, p = 10, 0.4
x = np.arange(binom.ppf(0.000001, n, p), binom.ppf(1, n, p))
fig = plt.figure(figsize = (10, 5))
ax = fig.add_subplot(1,2,1)
ax.plot(x, binom.pmf(x, n, p), 'bo', ms = 8, label = 'pmf binomial')
ax.vlines(x, 0,binom.pmf(x, n, p), colors = 'b', linewidth = 5, alpha = 0.5)
ax = fig.add_subplot(1,2,2)
ax.plot(x, binom.cdf(x, n, p), 'bo', ms = 8, label = 'cdf binomial')
ax.vlines(x, 0, binom.cdf(x, n, p), colors = 'b', linewidth = 5, alpha = 0.5)
fig.suptitle('Distribucion binomial')
plt.show()
```

```python
from scipy.stats import binom
#1- Prob de sacar 4 bolas rojas
binom.pmf(4, n = 10, p = 0.40)

#2- Prob de sacar al menos 4 rojas 
1 - binom.cdf(3, n=10, p=0.40)

#3- Prob de sacar menos de 3 rojas
binom.cdf(2, n=10, p=0.40)

#4- Valor esperado del nro de bolas rojas
binom.mean(n = 10, p = 0.40)

#5- Desv estandar 
binom.std(n = 10, p = 0.40)
```

```python
from scipy.stats import geom

# P(X=0) = (1-0.25)**0 * 0.25**1
geom.pmf(0, p = 0.25, loc = -1)
# P(X<=0) = 1 - (1 - 0.25) ** 1
geom.cdf(0, p = 0.25, loc = -1)
# Muestra aleatoria de tamaño 20
geom.rvs(p = 0.25, size = 20, loc = -1)
# Valor esperado
geom.mean(p = 0.25, loc = -1)
# Tambien calculada con los momentos (me devuelve la media y la varianza en este ejemplo)
geom.stats(p = 0.25, loc = -1, moments = 'mv')
```

```python
import matplotlib.pyplot as plt
import numpy as np
p = 0.25
x = np.arange(geom.ppf(0.00001, p), geom.ppf(0.99, p))
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (10,5))
ax1.plot(x, geom.pmf(x, p), 'ro')
ax1.set_title('PMF')
ax1.vlines(x, geom.pmf(x, p), ymax = 0, colors = 'g', linewidth = 3)

ax2.set_title('CDF')
ax2.plot(x, geom.cdf(x, p), 'ro')
ax2.vlines(x, geom.cdf(x, p), ymax = 0, colors = 'g', linewidth = 3)

fig.suptitle('Distribucion Geometrica', size = 16)
plt.show()
```
