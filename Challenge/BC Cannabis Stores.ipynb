```python
import scrapy
from scrapy.crawler import CrawlerProcess
import re
```

# Parte 1

```python
class Cannabis_spider(scrapy.Spider):
    name = 'Cannabis spider'
    
    def start_requests(self):
        urls = ['https://www.bccannabisstores.com/collections/cannabis-products?page='+str(i)+'&grid_list=list-view' for i in range(1,15)]
        for url in urls:
            yield scrapy.Request(url = url, callback = self.parse1)
        
    def parse1(self, response):
        producto = response.css('h2.productitem--title')
        links = producto.css('a::attr(href)').extract()
        for link in links:
            yield response.follow(url = link, callback = self.parse2)
            
    def parse2(self, response):
        nombres = response.css('h1.product-title>span::text').extract_first().strip()
        precio = response.css('span.money::text').extract_first().strip()
        sku = response.css('span.product-form--additional-info-sku::text').extract()[0][4:]
        presentacion = response.css('div.form-field-select-wrapper>select>option::attr(value)').extract()
        presentacion_ok = [re.findall(r'.+\D+.+', i) for i in presentacion if re.findall(r'.+\D+.+', i)]
        presentacion_final = [''.join(i) for i in presentacion_ok]
        thc = re.findall(r'\d+.+',response.css('div.product-details--thc-cbd span::text').extract()[0])[0]
        cbd = re.findall(r'\d+.+',response.css('div.product-details--thc-cbd span::text').extract()[1])[0]
        direccion = response.xpath('//link[@rel="canonical"]/@href').extract_first()
        
        for i in range(len(nombres)):
            resumen[sku] = [nombres, precio, presentacion_final, thc, cbd, direccion]
            
```

```python
resumen = {}
process = CrawlerProcess()
process.crawl(Cannabis_spider)
process.start()
```

```python
resumen
```

```python
import pandas as pd
productos_cannabis = pd.DataFrame().from_dict(resumen, orient = 'index', columns = ['Nombre', 'Precio', 'Presentacin', 'THC per container', \
                                                                  'CBD per container', 'URL'])
productos_cannabis.index.name = 'SKU'
productos_cannabis.head()
```

```python
productos_cannabis.to_csv('productos.csv')
```

```python
resumen_2 = {}
process = CrawlerProcess()
process.crawl(Accesorios_spider)
process.start()
```
